* 先生成母线的参数方程 on（x-z平面），然后旋转

```python
h = 1
```

```python
wp = WorkPlane()
face = wp.MoveTo(3*h,0).Rotate(90).Arc(h,90)\
.Line(2*h).Arc(h,-45).Line(2*h).Arc((np.sqrt(2)-1)*h,180).Line(2*h).Arc(h,-45).Close().Face()
```

```python
Draw(face)
```

```python
torus = face.Revolve(Axis((0,0,0), X), 360)
mesh = Mesh(OCCGeometry(torus).GenerateMesh(maxh=0.2))
```

```python
Draw(mesh)
```
