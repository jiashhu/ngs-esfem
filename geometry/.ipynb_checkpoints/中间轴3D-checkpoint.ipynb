* 先生成母线的参数方程 on（x-z平面），然后旋转

```python
def comput_MM(mesh):
    # M = 总质量 / 体积
    M = Integrate(1, mesh, element_wise=False)  

    # x, y, z 分量分别积分
    xc = Integrate(x, mesh, element_wise=False) / M
    yc = Integrate(y, mesh, element_wise=False) / M
    zc = Integrate(z, mesh, element_wise=False) / M

    x_c = np.array([xc, yc, zc])
    print(x_c)

    X = x - xc
    Y = y - yc
    Z = z - zc

    Ixx = Integrate(Y**2 + Z**2, mesh, element_wise=False)
    Iyy = Integrate(X**2 + Z**2, mesh, element_wise=False)
    Izz = Integrate(X**2 + Y**2, mesh, element_wise=False)
    Ixy = Integrate(-X*Y, mesh, element_wise=False)
    Ixz = Integrate(-X*Z, mesh, element_wise=False)
    Iyz = Integrate(-Y*Z, mesh, element_wise=False)

    I = np.array([[Ixx, Ixy, Ixz],
                  [Ixy, Iyy, Iyz],
                  [Ixz, Iyz, Izz]])


    eigvals, eigvecs = np.linalg.eigh(I)
    print("主惯量:", eigvals)
    print("对应方向:", eigvecs)
```

```python
h = 1
```

```python
wp = WorkPlane()
face = wp.MoveTo(3*h,0).Rotate(90).Arc(h,90)\
.Line(2*h).Arc(h,-45).Line(2*h).Arc((np.sqrt(2)-1)*h,180).Line(2*h).Arc(h,-45).Close().Face()
```

```python
Draw(face)
```

```python
torus = face.Revolve(Axis((0,0,0), X), 360)
mesh = Mesh(OCCGeometry(torus).GenerateMesh(maxh=0.2))
```

```python
Draw(mesh)
```

```python
comput_MM(mesh)
```
