```python
from ngsolve import *
from ngsolve.webgui import Draw
from netgen.occ import *
```

```python
h = 0.02
```

```python
square = Circle( (0,0), 1).Face() 
square_in = Circle( (0,0), 1-h).Face() 
band = square - square_in
```

```python
from geometry import *
```

```python
geo_in = geo_quadfoil_in_circ(1-h,bnd_in='bnd_in',bnd_out='bnd_out')
```

```python
shape = Glue([geo_in.shape, band])
```

```python
from ngsolve.webgui import Draw
Draw(shape);
```

```python
mesh = Mesh(OCCGeometry(shape, dim = 2).GenerateMesh(maxh=h))
```

```python
Draw(mesh)
```
